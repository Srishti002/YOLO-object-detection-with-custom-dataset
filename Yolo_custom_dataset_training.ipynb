```python
# YOLO on real time Video feed ( Bonus : we are just reminding user if they have taken or not their important thing when object is detected. If they have forgotten then a text will show fof the reminder)
from ultralytics import YOLO
import cv2
import numpy as np
import IPython.display as display
import threading
import time
from IPython.display import clear_output, Image
from PIL import Image as PILImage
import io

#important items that should be detected
IMPORTANT_ITEMS = ["phone", "laptop", "keys", "wallet"]

class ObjectDetection:
    def __init__(self, model_path):
        print(f"Loading model from {model_path}")
        self.model=YOLO(model_path)  
        self.cap=cv2.VideoCapture(0)
        self.running=False
        self.frame_rate=0.1  
        self.latest_frame = None
        self.display_handle = None
    
    def check_important_items(self, detected_names):  # Here we are detecting the object using and then if object detected is not in important Items list then we are prompting a text to not forget that items
        for item in IMPORTANT_ITEMS:
            if item not in detected_names:
                print(f"REMINDER!! Don't forget your {item} before leaving!!")
    
    def notify_detection(self, boxes, width, height):
        detected_names=[]
        for box in boxes:
            cls_id=int(box.cls[0].item())
            name=self.model.names[cls_id]
            detected_names.append(name)
            
        self.check_important_items(detected_names)
        return detected_names
    
    def display_thread(self):
        while self.running:
            if self.latest_frame is not None:
                
                rgb_frame=cv2.cvtColor(self.latest_frame, cv2.COLOR_BGR2RGB)
                pil_img=PILImage.fromarray(rgb_frame)
                
                buf=io.BytesIO()
                pil_img.save(buf, format='JPEG')
                img_bytes = buf.getvalue()
                
                if self.display_handle is None:
                    self.display_handle = display.display(display.Image(data=img_bytes), display_id=True)
                else:
                    self.display_handle.update(display.Image(data=img_bytes))
            
            time.sleep(self.frame_rate)
    
    def run(self):
        self.running = True
        
        display_thread = threading.Thread(target=self.display_thread)
        display_thread.daemon = True
        display_thread.start()
        
        print("Press 'q' in the output cell to stop")
        
        try:
            while self.running:
                ret, frame = self.cap.read()
                if not ret:
                    print("Failed to grab frame")
                    break
                
                height, width = frame.shape[:2]
                results = self.model(frame)
                
                if len(results) > 0 and len(results[0].boxes) > 0:
                    self.notify_detection(results[0].boxes, width, height)
                
                
                annotated_frame=results[0].plot()
                
                
                self.latest_frame=annotated_frame
                
                
                if cv2.waitKey(1) & 0xFF == ord('q'):
                    break
                
                
                time.sleep(0.01)
        
        finally:
            
            self.cap.release()
    
    def stop(self):
        self.running = False
        print("Stopping object detection")
```
